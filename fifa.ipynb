```python
import pandas as pd
import numpy as np
```

```python
players = pd.read_csv("players_22.csv")
```

```python
players
```

Let's list the features based on which we will cluster our players

* `overall` - overall rating of the player
* `potential` - potential rating of the player
* `value_eur` - total value of the player to the Club in Euros
* `wage_eur` - wage of the player in Euros
* `age` - age of the player

```python
features = ["overall", "potential", "wage_eur", "value_eur", "age"]
```

```python
# dropping null values from columns mentioned in features
players = players.dropna(subset=features)
# this will make sure that there will be no missing values in the columns that we are clustering
```

```python
data = players[features].copy()
```

```python
data
```

1. Scale the data
2. Initialize random centroids
3. Label each data points
4. Update centroids
5. Repeat steps 3 and 4 until centroids stop changing

## 1. Scaling the Data

```python
data = ((data - data.min()) / (data.max() - data.min())) * 9 + 1
```

```python
data.describe()
```

```python
data.head()
```

## 2. Intializing Random Centroids

```python
def random_centroid(data, k):
    centroids = []
    for i in range(k):
        centroid = data.apply(lambda x: float(x.sample()))
        centroids.append(centroid)
    return pd.concat(centroids, axis=1)
```

```python
centroids = random_centroid(data, 5)
```

```python
centroids
```

```python
def get_labels(data, centroids):
    distances = centroids.apply(lambda x: np.sqrt(((data - x) ** 2).sum(axis=1)))
    return distances.idxmin(axis=1)
```

```python
labels = get_labels(data, centroids)
```

```python
labels.value_counts()
```

```python
def new_centroids(data, labels, k):
    return data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
def plot_clusters(data, labels, centroids, iteration):
    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)
    centroids_2d = pca.transform(centroids.T)
    clear_output(wait=True)
    plt.title(f'Iteration {iteration}')
    plt.scatter(x=data_2d[:, 0], y=data_2d[:,1], c=labels)
    plt.show()
```

```python
max_iterations = 100
k = 3 # number of clusters

centroids = random_centroid(data, k)
old_centroids = pd.DataFrame()
iteration = 1

while iteration < max_iterations and not centroids.equals(old_centroids):
    old_centroids = centroids

    labels = get_labels(data, centroids)
    centroids = new_centroids(data, labels, k)
    plot_clusters(data, labels, centroids, iteration)
    iteration += 1
```

## Conclusion
Cluster 0 represets star players

Cluster 1 represents older players who have hit their potential.

Cluster 2 represents young players

K-means allows you to find patterns in  your data that you didn't know were there. This allowed us to categorize players automatically into different groups. 

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(3)
kmeans.fit(data)
```

```python
centroids = kmeans.cluster_centers_
```
